```python
import pandas as pd
```

```python
df = pd.read_csv('users.csv')
```

```python
print(df)
```

```python
import random
from faker import Faker
```

```python
fake = Faker()
user_names = [fake.name() for _ in range(len(df))]
```

```python
df['user_name'] = user_names
```

```python
print(df)
```

```python
order = ['userid', 'user_name', 'gender', 'age', 'ope', 'con', 'ext', 'agr', 'neu']
```

```python
df = df[order]
print(df)
```

```python
availability = [random.choice([True, False]) for _ in range(len(df))]
distance = distances = [random.uniform(0, 20) for _ in range(len(df))]
df['available'] = availability
df['distance'] = distances
```

```python
print(df)
```

```python
df.to_csv('modified_users.csv', index=False)
```
