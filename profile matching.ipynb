```python
selected = selected.drop(['available', 'index'], axis= 1)
print(selected)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
similarities = []
features = ['gender', 'age', 'ope', 'con', 'ext', 'agr', 'neu', 'distance']
selected_data = selected[features]
```

```python
shopper = selected.iloc[0]
```

```python
similarity_scores = cosine_similarity(selected_data, shopper[features].values.reshape(1, -1))
```

```python
similarity_df = pd.DataFrame({'Similarity': similarity_scores.flatten()})
```

```python
similarity_df = similarity_df.sort_values(by='Similarity', ascending=False)
```

```python
top_similar_rows = similarity_df.head(20)
```

```python
print(top_similar_rows)
```
