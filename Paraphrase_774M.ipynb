```
gpt2.download_gpt2(model_name="774M")
```

```
gpt2.mount_gdrive()
```

```
file_name = "msr_paraphrase_data_edited.txt"
```

```
gpt2.copy_file_from_gdrive(file_name)
```

```
sess = gpt2.start_tf_sess()

gpt2.finetune(sess,
              dataset=file_name,
              model_name='774M',
              steps=1000,
              restore_from='fresh',
              run_name='paraphrase774',
              print_every=10,
              sample_every=200,
              save_every=200)
gpt2.copy_checkpoint_to_gdrive(run_name='paraphrase774')
```

```
gpt2.copy_checkpoint_from_gdrive(run_name='paraphrase774')
```

```
sess = gpt2.start_tf_sess()
gpt2.load_gpt2(sess, run_name='paraphrase774')
```

```
text = gpt2.generate(sess,
                     run_name='paraphrase774', 
                     length= 50,
                     top_k= 40,
                     top_p= 0.9,
                     prefix='32.7 % of all households were made up of individuals and 15.7 % had someone living alone who was 65 years of age or older .',
                     return_as_list=True)[0]
```
